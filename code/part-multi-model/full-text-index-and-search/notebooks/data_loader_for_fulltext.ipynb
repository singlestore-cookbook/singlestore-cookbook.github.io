```python
from sqlalchemy import *

db_connection = create_engine(connection_url)
```

```python
with db_connection.begin() as conn:
    conn.execute(text("TRUNCATE TABLE articles;"))
```

```python
articles_df.to_sql(
    "articles",
    con = db_connection,
    if_exists = "append",
    index = False,
    chunksize = 1000
)
```
